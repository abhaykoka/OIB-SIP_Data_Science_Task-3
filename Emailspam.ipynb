```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python

stop_words=stopwords.words('english')
print(stop_words)
```

```python
df=pd.read_csv('spam11.csv')
stop_words=stopwords.words('english')
df
```

```python
data=df.where((pd.notnull(df)),'')
data
```

```python
data.describe()
data.head()
```

```python
data.loc[data['v1']=='spam','v1',]=0
data.loc[data['v1']=='ham','v1',]=1
x=data['v2']
y=data['v1']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(x_train)
```

```python

print(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features=feature_extraction.fit_transform(x_train)
x_test_features=feature_extraction.transform(x_test)
y_train_features=y_train.astype('int')
y_test_features=y_test.astype('int')
```

```python
model=LogisticRegression()
model.fit(x_train_features,y_train_features)
```

```python
pred=model.predict(x_train_features)
print(accuracy_score(y_train_features,pred)*100)
pred2=model.predict(x_test_features)
print(accuracy_score(y_test_features,pred2)*100)
```

```python

inp=["Hey!! Mr Abhay you have been elected as our next ceo"]
input_features=feature_extraction.transform(inp)
prediction=model.predict(input_features)
print(prediction)
```
